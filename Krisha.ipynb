```python
import time
import requests
import pandas as pd
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
from bs4 import BeautifulSoup
```

```python
c = 0
dlist = []
page = 1
while c < 200:
    print('page: {}'.format(page))
    query = "https://krisha.kz/arenda/kvartiry/?page={}".format(page)
    result = requests.get(query)
    src = result.content
    soup = BeautifulSoup(src, "lxml")
    for i in soup.find_all('div', {'class': 'a-card__inc'}):
        title = i.find('a', {'class': 'a-card__title'})
        price = i.find('div', {'class': 'a-card__price'})
        title1 = title.text.strip().split(',')
        rooms = title1[0]
        area = title1[1]
        area.strip('посуточно')
        area.strip('помесячно')
        area.strip('по часам')
        d = {
            'rooms': rooms,
            'area': area,
            'price': price.text.strip(),
        }
        d['title']=title.text.strip()
        d['price']=price.text.strip()
        dlist.append(d)
        print(title.text.strip(), price.text.strip(), '\n')
        c += 1
    page += 1
    
print("Total number of checked apartments: ", c)
print("Total number of checked pages: ", page)
print(dlist)
df = pd.DataFrame(dlist)
export_csv = df.to_csv('krisha_dataframe.csv', index=False)
print(df)
```
